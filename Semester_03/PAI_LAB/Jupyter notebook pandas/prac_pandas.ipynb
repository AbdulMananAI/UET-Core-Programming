```python
med_age = df['Age'].median() #find median of age 
df['Age'] = df['Age'].fillna(med_age) # fill null age entity with median value of age
df.isnull().sum()
```

```python
df['Embarked'].value_counts() # to know which value from embarked is most common to fll tnto missing
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
df.isnull().sum()
```

```python
df.drop(columns=['Cabin', 'Ticket']) #Cabin is mostly empty, and Ticket is just a random string of numbers that an AI can't easily learn from, drop them entirely 
```

```python
df.isnull().sum()
```

```python
# Add $5 to every Fare
df['Fare'] = df['Fare'] + 5

# Check the first few rows to see the updated Fares
df.head()
```

```python
# Extract all details for passenger 10
passenger_10 = df.loc[10]
passenger_10
```

```python
male_survivors = df[(df['Sex'] == 'male') & (df['Survived'] == 1)]

# Display the first 5 male survivors
male_survivors.head()
```

```python
# Filter for specific ports
c_or_q_ports = df[df['Embarked'].isin(['C', 'Q'])]

c_or_q_ports.head()
```

```python
# Find names containing "Miss." unmarried
miss_passengers = df[df['Name'].str.contains('Miss.', na=False)]

miss_passengers.head()
```

```python
# Create FamilySize column
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Check the new column
df[['Name', 'SibSp', 'Parch', 'FamilySize']].head()
```

```python
# Create Fare_Per_Person column
df['Fare_Per_Person'] = df['Fare'] / df['FamilySize']

df[['Name', 'Fare', 'FamilySize', 'Fare_Per_Person']].head()
```
